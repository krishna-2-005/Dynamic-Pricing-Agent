# Exploratory Data Analysis (EDA) for Dynamic Pricing
**Dataset:** retail_price_dataset.csv
**Goal:** Understand the data, uncover patterns, and guide feature engineering and model selection.

## 1. Setup and Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# Load data
df = pd.read_csv('../data/retail_price_dataset.csv')
print('Dataset shape:', df.shape)
df.head()
```

## 2. Basic Information and Data Types

```python
df.info()

# Parse date
df['date'] = pd.to_datetime(df['month_year'], format='%d-%m-%Y')
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
```

## 3. Summary Statistics

```python
df.describe(include='all')
```

## 4. Univariate Analysis

```python
# Quantity distribution
fig, axes = plt.subplots(1, 2, figsize=(14,4))
sns.histplot(df['qty'], bins=50, ax=axes[0])
axes[0].set_title('Distribution of Quantity Sold')
sns.boxplot(x=df['qty'], ax=axes[1])
axes[1].set_title('Boxplot of Quantity Sold')
plt.tight_layout()
plt.show()
```

```python
# Unit price distribution
fig, axes = plt.subplots(1, 2, figsize=(14,4))
sns.histplot(df['unit_price'], bins=50, ax=axes[0])
axes[0].set_title('Distribution of Unit Price')
sns.boxplot(x=df['unit_price'], ax=axes[1])
axes[1].set_title('Boxplot of Unit Price')
plt.tight_layout()
plt.show()
```

```python
# Product score distribution
sns.countplot(x='product_score', data=df)
plt.title('Product Score Distribution')
plt.show()
```

## 5. Bivariate Analysis

```python
# Price vs Quantity
plt.figure(figsize=(10,6))
sns.scatterplot(x='unit_price', y='qty', data=df, alpha=0.5)
plt.title('Price vs Quantity Sold')
plt.show()
```

```python
# Correlation heatmap
num_cols = df.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(14,10))
sns.heatmap(df[num_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

## 6. Time Series Analysis

```python
# Monthly sales trend
monthly_sales = df.groupby('date')['qty'].sum().reset_index()
monthly_sales = monthly_sales.sort_values('date')

plt.figure(figsize=(14,5))
plt.plot(monthly_sales['date'], monthly_sales['qty'], marker='o')
plt.title('Monthly Total Quantity Sold')
plt.xlabel('Date')
plt.ylabel('Total Quantity')
plt.grid(True)
plt.show()
```

## 7. Competitor Price Analysis

```python
# Competitor price distribution
comp_prices = df[['comp_1','comp_2','comp_3']].stack().reset_index(drop=True)
comp_prices = comp_prices[comp_prices > 0]

plt.figure(figsize=(10,5))
sns.histplot(comp_prices, bins=50)
plt.title('Distribution of Competitor Prices')
plt.xlabel('Price')
plt.show()
```

## 8. Missing Values and Outliers

```python
# Missing values
df.isnull().sum()

# Outlier capping
Q99 = df['qty'].quantile(0.99)
df_no_outliers = df[df['qty'] <= Q99]
print(f'Rows before: {len(df)}, after capping 99th percentile: {len(df_no_outliers)}')
```

## 10. Summary of Insights
- Quantity sold is highly skewed; log transformation may help.
- Price alone is a weak predictor; competitor prices and product attributes matter.
- Seasonality exists in monthly patterns.
- Different product categories show different demand levels.
- Customer count correlates with quantity sold.
- Product ratings may influence willingness to pay.
